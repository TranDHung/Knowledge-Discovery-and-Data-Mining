```python
import pandas as pd
#Read .csv file
df = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
df.head()
```

```python
#Convert dataframe to list of transaction
transactions = df.stack().groupby(level=0).apply(list).tolist()
#Get first 100 transactions
transactions = transactions[:100]
transactions[:10]
```

```python
def Eclat(P, minsup, F):
  """
  Eclat algorithm
  Input: P (dict) 1-itemset with tidsets; minsup (int) as minium support;
         F (list) is result as list of frequent itemset
  Output: list of frequent itemset.
  """
  for Xa, tXa in P.items():
    F.append([(Xa), len(tXa)])
    Pa = {}
    for Xb, tXb in P.items():
      if Xb > Xa:
        Xab = tuple(set([Xa]) | set([Xb]))
        tXab = list(set(tXa) & set(tXb))
        if len(tXab) >= minsup:
          Pa[Xab] = tXab
    if Pa!={}:
      Eclat(Pa, minsup, F)
```

```python
#Create P 1-itemset
P = {}
count_tid = 1
for trans in transactions:
  for item in trans:
    if item in P:
      P[item].append(count_tid) 
    else:
      P[item] = [count_tid]
  count_tid+=1

P.items()
```

```python
F = [] #Initial F
minsup = 5 #Initial minsup
#remove items with sup < minsup
for key, value in list(P.items()):
  if (len(value) < minsup):
    del P[key]
```

```python
Eclat(P,minsup,F)
F
```
